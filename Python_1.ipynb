```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Dim_customer=pd.read_csv('/content/Dim_customer.csv')
Dim_employee=pd.read_csv('/content/Dim_employee.csv')
Dim_playlist=pd.read_csv('/content/Dim_playlist.csv')
Dim_track=pd.read_csv('/content/Dim_track.csv')
Dim_currency=pd.read_csv('/content/Dim_currency.csv')
Fact_invoice=pd.read_csv('/content/Fact_invoice.csv')
Fact_invoiceline=pd.read_csv('/content/Fact_invoiceline.csv')
```

```python
track_top_5=Dim_track.groupby("artist_name")["trackid"].count().sort_values(ascending=False).head(5) #Get the track count by artist
track_top_5
```

```python
album_top_5=Dim_track.groupby("artist_name")["albumid"].nunique().sort_values(ascending=False).head(5) #Get the album count by artist
album_top_5
```

```python
genre_top_5=Dim_track.groupby("genre_name")["track_name"].count().sort_values(ascending=False).head(5) #The track count by genre
genre_top_5
```

```python
fig, axs = plt.subplots(3, 1, figsize=(5, 12)) # Creates a figure with 3 rows and 1 column of subplots.

axs[0].bar(genre_top_5.index, genre_top_5.values, color='orange') #Draws a bar chart using Genre as the labels (x-axis) and track count as the heights of the bars (y-axis).
axs[0].set_title("Top 5 Genres with more Tracks")
axs[0].set_xlabel("Genre")
axs[0].set_ylabel("Tracks")
axs[0].tick_params(axis='x', rotation=45)

axs[1].bar(album_top_5.index, album_top_5.values, color='purple')#Draws a bar chart using Artists as the labels (x-axis) and album count as the heights of the bars (y-axis).
axs[1].set_title("Top 5 Artists with more Albums")
axs[1].set_xlabel("Artist")
axs[1].set_ylabel("Albums")
axs[1].tick_params(axis='x', rotation=45)

axs[2].bar(track_top_5.index, track_top_5.values, color='pink')#Draws a bar chart using Artists as the labels (x-axis) and track count as the heights of the bars (y-axis).
axs[2].set_title("Top 5 Artists with more Tracks")
axs[2].set_xlabel("Artist")
axs[2].set_ylabel("Tracks")
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
